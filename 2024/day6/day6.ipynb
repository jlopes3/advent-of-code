## Part 1

```python
from functools import reduce

map: list[list[str]] = []
with open("input_day6.txt", "r") as file:
    line: str = file.readline()
    while line:
        map.append(list(line.strip()))
        line = file.readline()
        
def search_for_guard(map: list[list[str]]):
    height: int = len(map)
    width: int = len(map[0])
    for i in range(height):
        for j in range(width):
            if map[i][j] in [">", "<", "^", "v"]:
                return i , j
    return -1, -1


def guard_one_step(map, guard_loc, guard_dir):
    guard_dir_dict: dict[str, tuple[int, int]] = {">": (0, 1), "<": (0, -1), "^": (-1, 0), "v": (1, 0)}
    guard_turn_dict: dict[str, str] = {">": "v", "<": "^", "^": ">", "v": "<"}
    i_inc: int = guard_dir_dict[guard_dir][0]
    j_inc: int = guard_dir_dict[guard_dir][1]
    height: int = len(map)
    width: int = len(map[0])
    if (guard_loc[0] + i_inc < 0) or (guard_loc[0] + i_inc >= height) or (guard_loc[1] + j_inc < 0) or (guard_loc[1] + j_inc >= width):
        map[guard_loc[0]][guard_loc[1]] = "X"
        return map, (guard_loc[0] + i_inc, guard_loc[1] + j_inc), "_"
    if map[guard_loc[0] + i_inc][guard_loc[1] + j_inc] != "#":
        map[guard_loc[0] + i_inc][guard_loc[1] + j_inc] = guard_dir
        map[guard_loc[0]][guard_loc[1]] = "X"
        return map, (guard_loc[0] + i_inc, guard_loc[1] + j_inc), guard_dir
    else:
        map[guard_loc[0]][guard_loc[1]] = guard_turn_dict[guard_dir]
        return map, guard_loc, guard_turn_dict[guard_dir]
    

def run_sim(map):
    guard_loc: tuple[int, int] = search_for_guard(map)
    guard_dir: str = map[guard_loc[0]][guard_loc[1]]   
    while guard_dir != "_":
        map, guard_loc, guard_dir = guard_one_step(map, guard_loc, guard_dir)
    return map


def count_xs(map: list[list[str]]) -> int:
    height: int = len(map)
    width: int = len(map[0])
    counter: int = 0
    for i in range(height):
        for j in range(width):
            if map[i][j] == "X":
                counter += 1
    return counter  

map = run_sim(map)
xs: int = count_xs(map)
print("X's: " + str(xs))

for el in map:
    line_str: str = reduce(lambda x, y: str(x) + str(y), el)
    print(line_str)
```

## Part 2

```python

map: list[list[str]] = []
with open("input_day6.txt", "r") as file:
    line: str = file.readline()
    while line:
        map.append(list(line.strip()))
        line = file.readline()
        
height: int = len(map)
width: int = len(map[0])
        
def search_for_guard(map: list[list[str]]):
    for i in range(height):
        for j in range(width):
            if map[i][j] in [">", "<", "^", "v"]:
                return i , j
    return -1, -1

r, c = search_for_guard(map)

def loops(map:list[list[str]], r: int, c: int) -> bool:
    dr: int = -1
    dc: int = 0

    seen: set[tuple[int, int, int, int]] = set()

    while True: 
        seen.add((r, c, dr, dc))
        if (r + dr < 0) or (r + dr >= height) or (c + dc < 0) or (c + dc >= width):
            return False
        if map[r + dr][c + dc] == "#":
            dc, dr = -dr, dc
        else:
            r += dr
            c += dc
        if (r, c, dr, dc) in seen:
            return True
        

cycle_counter: int = 0
for i, j in seen:
    if map[i][j] != ".":
        continue
    map[i][j] = "#"
    if loops(map, r, c):
        cycle_counter += 1
    map[i][j] = "."
        
print(cycle_counter)
print(seen)    
```
