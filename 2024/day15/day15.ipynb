```python
lines: list[str] = []
with open("input_day15.txt", "r") as file:
    board_complete: bool = False
    line: str = file.readline()
    while line:
        lines.append(line.strip())
        line = file.readline()

i: int = 1
commands: str = ""
while len(lines[-i]) > 0:
    commands = lines[-i] + commands
    i += 1
board: list[str] = lines[0:-i]
WIDTH: int = len(board[0])
HEIGHT: int = len(board)

walls: set[tuple[int, int]] = set()
boxes: set[tuple[int, int]] = set()
robot: tuple[int, int] = (-112358, -112358)
for j in range(WIDTH):
    for i in range(HEIGHT):
        if board[i][j] == "#":
            walls.add((i, j))
        elif board[i][j] == "O":
            boxes.add((i, j))
        elif board[i][j] == "@":
            robot = (i, j)

def print_board(walls: set[tuple[int, int]], boxes: set[tuple[int, int]], robot: tuple[int, int], height: int, width: int) -> None:
    for i in range(height):
        for j in range(width):
            if (i, j) in walls:
                print("#", end = " ")
            elif (i, j) in boxes:
                print("O", end = " ")
            elif (i, j) == robot:
                print("@", end=" ")
            else:
                print(".", end=" ")
        print()

def move_possible(command: str, walls: set[tuple[int, int]], boxes: set[tuple[int, int]], robot: tuple[int, int], height: int, width: int) -> bool:
    c: int
    r: int
    r, c = robot
    if command == "<":
        for j in range(c - 1, -1, -1):
            if (r, j) in walls:
                return False
            elif (r, j) not in boxes:
                return True
    elif command == ">":
        for j in range(c+ 1, WIDTH):
            if (r, j) in walls:
                return False
            elif (r, j) not in boxes:
                return True
    elif command == "^":
        for i in range(r - 1, -1, -1):
            if (i, c) in walls:
                return False
            elif (i, c) not in boxes:
                return True
    elif command == "v":
        for i in range(r + 1, HEIGHT):
            if (i, c) in walls:
                return False
            elif (i, c) not in boxes:
                return True
    return False 

def make_move(command: str, walls: set[tuple[int, int]], boxes: set[tuple[int, int]], robot: tuple[int, int], height: int, width: int) -> tuple[int, int]:
    c: int
    r: int
    r, c = robot
    add_box: bool = False
    next_box_spot: tuple[int, int] = (-100, -100)
    new_robot: tuple[int, int] = robot
    if command == "<":
        new_robot = (robot[0] , robot[1] - 1)
        for j in range(c - 1, -1, -1):
            if (r, j) in boxes:
                add_box = True
            elif (r, j) not in boxes:
                next_box_spot = (r, j)
                break
    elif command == ">":
        new_robot = (robot[0] , robot[1] + 1)
        for j in range(c+ 1, WIDTH):
            if (r, j) in boxes:
                add_box = True
            elif (r, j) not in boxes:
                next_box_spot = (r, j)
                break
    elif command == "^":
        new_robot = (robot[0] - 1, robot[1])
        for i in range(r - 1, -1, -1):
            if (i, c) in boxes:
                add_box = True
            elif (i, c) not in boxes:
                next_box_spot = (i, c)
                break
    elif command == "v":
        new_robot = (robot[0] + 1, robot[1])
        for i in range(r + 1, HEIGHT):
            if (i, c) in boxes:
                add_box = True
            elif (i, c) not in boxes:
                next_box_spot = (i, c)
                break
    
    if add_box:
        boxes.add(next_box_spot)
    if new_robot in boxes: 
        boxes.remove(new_robot)
    return new_robot 



for command in commands:
    if move_possible(command ,walls, boxes, robot, HEIGHT, WIDTH):
        robot = make_move(command ,walls, boxes, robot, HEIGHT, WIDTH)
    
gps_sum: int = 0
for el in boxes:
    gps_sum += el[0] * 100 + el[1]
print(gps_sum)        
    
            
```
